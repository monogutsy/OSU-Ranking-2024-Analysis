## **OSU STANDARD RANKING 2024**

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_style("darkgrid")
```

```python
#importing the CSV file
file_path = "OSU_Standard_Ranking_2024.csv" 
#reading the CSV file using pandas
osu_data = pd.read_csv(file_path)
```

```python
#data information of the file
osu_data
```

```python
#removing (%) percentage and (,) coma to be able to do math
def clean_currency(x):
    if isinstance(x, str):
        return x.replace(',', '').replace('#', '').replace('%', '').replace('\u202f', '').replace(' ', '')
    return x
```

```python
#list of name of the columns header
osu_data.columns
```

```python
#editing columns name
v1 = ['Rank', 'Performance', 'Country Ranking', 'Accuracy', 
                 'Play Count', 'Ranked Score', 'Total Score', 'SS', 'S', 'A']
for col in v1:
    osu_data[col] = osu_data[col].apply(clean_currency).astype(float)
```

```python
#final information of the CSV file
display(osu_data.head())
osu_data.info()
```

### **2. IS THERE A CORRELATION BETWEEN ACCURACY AND PERFORMANCE (PP)?**

<p>

    <br>

    <br><br>
    <b>CONCLUSION:</b>
    <br>
    Scientific research shows that you can't be both perfectly fast and perfectly accurate at the same time. Top players know this, so they intentionally play maps that are 'too fast' for them. Even if their accuracy drops a little, the map is so hard that they still get way more points than if they played it safe.

### **3. HOW DOES PLAYTIME CORRELATE WITH PLAYER LEVEL?**

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=osu_data, x='Level', y='Playtime', color='coral')
plt.title('Trend of Playtime Required to Reach Higher Levels', fontsize=16)
plt.xlabel('Player Level')
plt.ylabel('Playtime (Hours)')
plt.show()
```

<p>
    <b>INSIGHTS:</b>
    <br>
    The line shoots up because leveling up gets harder, not easier. You can breeze through the early levels in a few hours, but once you reach the high levels, you have to play for hundreds of hours just to gain one single level.
    <br><br>
    <b>CONCLUSION:</b>
    <br>
    It follows a rule called the 'Power Law.' At the start, you improve very quickly with little effort. But later on, you have to work ten times harder just to make a tiny bit of progress. Game designers do this on purpose so you always have a distant goal to chase.
    <br><br>
    <b>BIBLIOGRAPHY:</b>
    <br>
    Newell, A., & Rosenbloom, P. S. (1981). "Mechanisms of skill acquisition and the law of practice". Cognitive skills and their acquisition.
</p>

<p>

If a dot is low, that player has great accuracy but isn't getting enough points for it, likely because they are sticking to easier maps.
    <br><br>
